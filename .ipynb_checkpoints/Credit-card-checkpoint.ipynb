```python
#Copia
df1 = df.copy()
```

```python
# Deletar a coluna ID
del df1['ID']; df1.head()
```

```python
# Colunas
df1.columns
```

```python
# Alterar o nome da coluna de classe para inadimplente
df2 = df1.rename(columns={'default payment next month':'inadimplente'}); df2.head()
```

```python
# Verificar se há valores nulos
df2.isnull().sum()
```

```python
# Renomear colunas
colunas = {'SEX':'Genero', 'EDUCATION':'Escolaridade', 'MARRIAGE':'Estado_Civil', 'AGE':'Idade'}
```

```python
df2.rename(columns=colunas, inplace=True); df2.head()
```

```python
# Gêneros(2-F, 1-M)
df2['Genero'].unique()
```

```python
df3 = df2.copy()
```

```python
# Amostragem estratificada
from sklearn.model_selection import train_test_split
```

```python
# Treino e teste
prev_treinamento, prev_teste, class_treinamento, class_teste = train_test_split(df3[['PAY_0', 'PAY_3', 'PAY_AMT1', 'PAY_AMT2', 'PAY_5', 'BILL_AMT1']], df3['inadimplente'],
                              test_size = 0.25, stratify = df3['inadimplente'])
```

```python
# Modelo V1 - Naive Bayes
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
previsao = model.fit(prev_treinamento, class_treinamento)
```

```python
# Teste
prev = previsao.predict([[0,1,1100,1500,0,350], [0, 0, 1000, 1300, 0, 420], [0, 0, 1200, 1150, 0, 280]])
prev
```
